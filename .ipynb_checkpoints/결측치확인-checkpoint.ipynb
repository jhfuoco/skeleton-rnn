```python
import os
import pandas as pd

def check_missing_values_in_folder(folder_path):
    missing_values = {}
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".csv"):
            filepath = os.path.join(folder_path, filename)
            df = pd.read_csv(filepath)
            
            # 결측치 확인
            missing_count = df.isnull().sum()
            
            # 결측치가 있는 컬럼만 필터링
            missing_count = missing_count[missing_count > 0]
            
            if not missing_count.empty:  # 결측치가 있는 경우만 저장
                missing_values[filename] = missing_count
            else:
                missing_values[filename] = "No missing values"
                
    return missing_values

folder_path = "C:/jihoon/walking_dataset_withcsv/parkinson"  # 폴더 경로를 여기에 지정하세요.
missing_values_dict = check_missing_values_in_folder(folder_path)

# 결과 출력
for filename, missing_count in missing_values_dict.items():
    print(f"In {filename}:\n{missing_count}\n{'-'*40}")

```
